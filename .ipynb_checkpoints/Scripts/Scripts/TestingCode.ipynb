## Calculate_gower_distance

```python
from sklearn.neighbors import DistanceMetric
import pandas as pd
import numpy as np

def calculate_gower_distance(df):

    variable_distances = []

    for col in range(df.shape[1]):
        feature = df.iloc[:,[col]]
        if feature.dtypes.values == object:
            feature_dist = DistanceMetric.get_metric('dice').pairwise(pd.get_dummies(feature, drop_first=True))
        else:
            feature_dist = DistanceMetric.get_metric('manhattan').pairwise(feature) / max(np.ptp(feature.values),1)
            variable_distances.append(feature_dist)
    return np.array(variable_distances).mean(0)
```

```python
import numpy as np
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import pairwise_distances

# Generar datos de muestra
num_samples = 100
num_features = 5

# Generar datos aleatorios
data = np.random.rand(num_samples, num_features)

# Calcular la matriz de distancias
dist_matrix = pairwise_distances(data, metric='euclidean')

# Definir el n√∫mero de clusters
num_clusters = 3

# Definir una semilla aleatoria para reproducibilidad
random_state = 42

# Aplicar KMedoids
kmedoids = KMedoids(n_clusters=num_clusters, 
                    metric='precomputed',
                    method='pam', init='build', 
                    max_iter=300, 
                    random_state=random_state)

# Ajustar el modelo y obtener las etiquetas de los clusters
clusters = kmedoids.fit(dist_matrix).labels_

print(clusters)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Graficar los datos de muestra
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'g', 'b']  # Colores para los clusters

for i in range(num_clusters):
    ax.scatter(data[clusters == i, 0], data[clusters == i, 1], data[clusters == i, 2], c=colors[i], label=f'Cluster {i}')

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('KMedoids Clustering')

# Agregar una leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Graficar los datos de muestra
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'g', 'b']  # Colores para los clusters

for i in range(num_clusters):
    # Plotear los puntos del cluster
    ax.scatter(data[clusters == i, 0], data[clusters == i, 1], data[clusters == i, 2], c=colors[i], label=f'Cluster {i}')
    
    # Agregar esferas para resaltar el cluster
    cluster_center = np.median(data[clusters == i], axis=0)  # Calcular el centro del cluster
    cluster_radius = np.max(pairwise_distances(data[clusters == i], [cluster_center]))  # Calcular el radio del cluster
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = cluster_radius * np.outer(np.cos(u), np.sin(v)) + cluster_center[0]
    y = cluster_radius * np.outer(np.sin(u), np.sin(v)) + cluster_center[1]
    z = cluster_radius * np.outer(np.ones(np.size(u)), np.cos(v)) + cluster_center[2]
    ax.scatter(x, y, z, color=colors[i], alpha=0.1)  # Agregar la esfera al plot

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('KMedoids Clustering')

# Agregar una leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Graficar los datos de muestra
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'g', 'b']  # Colores para los clusters

# Almacenar √≠ndices de los outliers
outlier_indices = []

for i in range(num_clusters):
    # Plotear los puntos del cluster
    cluster_data = data[clusters == i]
    ax.scatter(cluster_data[:, 0], cluster_data[:, 1], cluster_data[:, 2], c=colors[i], label=f'Cluster {i}')
    
    # Calcular el centro y radio del cluster
    cluster_center = np.median(cluster_data, axis=0)
    cluster_radius = np.max(pairwise_distances(cluster_data, [cluster_center]))
    
    # Filtrar outliers
    cluster_outliers = np.array([idx for idx, point in enumerate(data) if np.linalg.norm(point - cluster_center) > cluster_radius])
    outlier_indices.extend(cluster_outliers)
    ax.scatter(data[cluster_outliers][:, 0], data[cluster_outliers][:, 1], data[cluster_outliers][:, 2], c='black', marker='x', label=f'Outliers {i}')

    # Agregar esferas para resaltar el cluster
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = cluster_radius * np.outer(np.cos(u), np.sin(v)) + cluster_center[0]
    y = cluster_radius * np.outer(np.sin(u), np.sin(v)) + cluster_center[1]
    z = cluster_radius * np.outer(np.ones(np.size(u)), np.cos(v)) + cluster_center[2]
    ax.plot_surface(x, y, z, color=colors[i], alpha=0.1)  # Agregar la esfera al plot

# Remover duplicados en outlier_indices
outlier_indices = list(set(outlier_indices))

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('KMedoids Clustering')

# Agregar una leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()

print("√çndices de outliers:", outlier_indices)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos para 14 d√≠as de visitas estables
dias_estables = np.arange(1, 15)
visitas_estables = np.random.normal(loc=100, scale=5, size=14)

# Generar un pico repentino en el d√≠a 15
dia_pico = 15
visitas_pico = 1000

# Generar datos para 15 d√≠as despu√©s del pico
dias_despues_pico = np.arange(dia_pico + 1, dia_pico + 16)
visitas_despues_pico = np.random.normal(loc=100, scale=5, size=15)

# Combinar todos los datos
dias = np.concatenate((dias_estables, [dia_pico], dias_despues_pico))
visitas = np.concatenate((visitas_estables, [visitas_pico], visitas_despues_pico))

# Graficar los datos
plt.figure(figsize=(10, 5))
plt.plot(dias, visitas, marker='o', linestyle='-')
plt.title('Red Empresa')
plt.xlabel('D√≠a')
plt.ylabel('Packets')
plt.grid(True)
plt.xticks(np.arange(1, 31, step=1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Funci√≥n para generar direcciones IP aleatorias en formato IPv4
def generate_ip():
    return ".".join(str(np.random.randint(0,256)) for _ in range(4))

# Funci√≥n para generar n√∫meros de puerto aleatorios
def generate_port():
    return np.random.randint(1, 65536)

# Funci√≥n para generar un protocolo aleatorio (TCP/UDP)
def generate_protocol():
    return np.random.choice(["TCP", "UDP"])

# Funci√≥n para generar el n√∫mero de warnings aleatorios
def generate_warnings():
    return np.random.randint(0, 10)

# Generar datos aleatorios
num_records = 100
data = {
    "IP": [generate_ip() for _ in range(num_records)],
    "Port": [generate_port() for _ in range(num_records)],
    "Protocol": [generate_protocol() for _ in range(num_records)],
    "Warnings": [generate_warnings() for _ in range(num_records)]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Mostrar los primeros registros
print(df.head())
```
